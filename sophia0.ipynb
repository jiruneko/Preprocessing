```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris)
```

```python
print("特徴料の名前=", iris.feature_names)
print("分類の名前=", iris.target_names)
print("分類の値=", iris.target)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data)
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 3種類の品種を、別々のデータフレームに分ける
# targetの値が0ならdf0に、1ならdf1に、2ならdf2に入れる
df0 = df[iris.target==0]
df1 = df[iris.target==1]
df2 = df[iris.target==2]

# 3種類の品種を、ヒストグラムで色分けして描画
# 「がくの幅」で、ヒストグラムを描画
plt.figure(figsize=(5, 5))
xx = "sepal width (cm)"
df0[xx].hist(color="b",alpha=0.5)
df1[xx].hist(color="r",alpha=0.5)
df2[xx].hist(color="g",alpha=0.5)
plt.xlabel(xx)
plt.show()
```

```python
xx = "sepal width (cm)"
yy = "sepal length (cm)"
plt.figure(figsize=(5, 5))
plt.scatter(df0[xx], df0[yy], color="b", alpha=0.5)
plt.scatter(df1[xx], df1[yy], color="r", alpha=0.5)
plt.scatter(df2[xx], df2[yy], color="g", alpha=0.5)
plt.xlabel(xx)
plt.ylabel(yy)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 「がくの幅」と「がくの長さ」と「花びらの長さ」で、3D散布図
xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# df0, df1, df2 をそれぞれ異なるデータフレームにする
ax.scatter(df0[xx], df0[yy], df0[zz], color="b", label='Class 0')  # 青
ax.scatter(df1[xx], df1[yy], df1[zz], color="r", label='Class 1')  # 赤
ax.scatter(df2[xx], df2[yy], df2[zz], color="g", label='Class 2')  # 緑

ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)

# 凡例を表示
plt.legend()
plt.show()
```

```python
ax = Axes3D(plt.figure(figsize=(5, 5)))
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df0[xx], df0[yy], df0[zz], color="b")
ax.scatter(df1[xx], df1[yy], df1[zz], color="r")
ax.scatter(df2[xx], df2[yy], df2[zz], color="g")
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(0, 240)
plt.show()
```

```python
ax = Axes3D(plt.figure(figsize=(5, 5)))

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df0[xx], df0[zz], color="b")
ax.scatter(df1[xx], df1[zz], color="r")
ax.scatter(df2[xx], df2[zz], color="g")
ax.set_xlabel(xx)
ax.set_ylabel(zz)
ax.set_zlabel(zz)
ax.view_init(90, 270)
plt.show()
```

```python
from sklearn.datasets import make_blobs
import pandas as pd

X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=2,
    cluster_std=1,
    n_samples=300,
)

df = pd.DataFrame(X)
df["target"] = y
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 分類によって、別々のデータフレームに分ける
df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
# 分類0は青、分類1は赤で、散布図を描画
plt.figure(figsize=(5, 5))
plt.scatter(df0[0], df0[1], color="b", alpha=0.5)
plt.scatter(df1[0], df1[1], color="r", alpha=0.5)
plt.grid()
plt.show()
```

```python
X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=3,
    cluster_std=1,
    n_samples=300)

# 特徴量（X）でデータフレームを作り、分類（y）をtargetの列として追加
df = pd.DataFrame(X)
df["target"] = y
# 分類によって、別々のデータフレームに分ける
df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]
# 分類0は青、分類1は赤、分類2は緑で、散布図を編集
plt.figure(figsize=(5, 5))
plt.scatter(df0[0], df0[1], color="b", alpha=0.5)
plt.scatter(df1[0], df1[1], color="r", alpha=0.5)
plt.scatter(df2[0], df2[1], color="g", alpha=0.5)
plt.grid()
plt.show()
```

```python
X, y = make_blobs(
    random_state=3,
    n_features=2,
    centers=5,
    cluster_std=1,
    n_samples=300)
# 特徴量（X）でデータフレームを作り、分類（y）をtargetの列として追加
df = pd.DataFrame(X)
df["target"] = y
# 分類によって、別々のデータフレームに分ける
df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]
df3 = df[df["target"]==3]
df4 = df[df["target"]==4]
# 分類0は青、1は赤、2は緑、3はマゼンタ、4はシアンで、散布図を描画
plt.figure(figsize=(5, 5))
plt.scatter(df0[0], df0[1], color="b", alpha=0.5)
plt.scatter(df1[0], df1[1], color="r", alpha=0.5)
plt.scatter(df2[0], df2[1], color="g", alpha=0.5)
plt.scatter(df3[0], df3[1], color="m", alpha=0.5)
plt.scatter(df4[0], df4[1], color="c", alpha=0.5)
plt.grid()
plt.show()
```
